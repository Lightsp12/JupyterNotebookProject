```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline





```

```python
df = pd.read_csv(r'C:\Users\jesus\Downloads\movies.csv')
```

```python
df.head()
```

```python
#Make copy out of the original Dataset and view the rows X cloumns numbers. 
df_modified = df.copy()
df_modified.head()
```

```python
# Viewing all the rows in cloumns
pd.set_option('display.max_rows', None)
```

```python
# Checking for missing values
for col in df_modified.columns:
    pct_missing = np.mean(df_modified[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
df_modified.dtypes
```

```python
df_modified['yearcorrect'] = df_modified['released'].astype(str).str[4:]
df_modified.head()
```

```python
# Transforming the correct year value in yearcorrect cloumn 
new_df = df_modified.replace({'yearcorrect': '/'}, '', regex =True)
new_df.head()
```

```python
new_df.sort_values(by=['gross'], inplace = False , ascending = False)
```

```python
# Dropin any duplicates from Comany cloumns
new_df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
new_df.head()
```

```python
# Scatter plot with budget vs genre and budget vs gross

plt.scatter(x=new_df['budget'], y=new_df['genre'])

plt.title('Budget vs Genre')

plt.xlabel( ' Movies Genre')
plt.ylabel(' Budget for Film')
plt.show()

```

```python
# Scatter plot with budget vs gross

plt.scatter(x=new_df['budget'], y=new_df['gross'] )

plt.title('Budget vs Gross Earnings')

plt.xlabel( ' Gross Earnings')
plt.ylabel(' Budget for Film')
plt.show()
```

```python
#plot Budget vs Gross applying seaborn 
sns.regplot (x='budget', y='gross', data= new_df, scatter_kws={"color": "green"} )
```

```python
#pearson, kendall, spearman - correlation types
# Identifation correlation in the dataset(numerical values only)
new_df.corr(method='pearson')
```

```python
correlation_matrix = new_df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel( ' Movie Features')
plt.ylabel(' Movie Features')

plt.show()
```

```python
# making the object data type numerical for better correlation porposied
df_numerized = new_df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
       
```
